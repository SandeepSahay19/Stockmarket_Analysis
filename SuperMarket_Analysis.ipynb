```python
df["Time"] = pd.to_datetime(df["Time"], format="%H:%M:%S")

# Group the data by hour and calculate the total quantity sold in each hour
hourly_sales = df.groupby(df["Time"].dt.hour)["Quantity"].sum()

# Find the hour with the highest total quantity sold
most_common_hour = hourly_sales.idxmax()

# Find the time range for the most common hour
start_time = f"{most_common_hour:02}:00:00"
end_time = f"{most_common_hour:02}:59:59"

print(f"The most shopping is done between {start_time} and {end_time}.")

```

### Most shoppings were done between 7 pm to 8 pm
